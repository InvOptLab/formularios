```python
import os
import json
```

```python
# Pasta onde estão os JSONs
pasta = 'formularios'
resultado = {"versao": "3.0", "formularios": {}, "disciplinas": {}, "docentes": [], "saldos": {}}
```

# Formulários

```python
# Percorre todos os arquivos .json da pasta
for arquivo in os.listdir(pasta):
    if arquivo.endswith('.json'):
        caminho = os.path.join(pasta, arquivo)
        with open(caminho, 'r', encoding='utf-8') as f:
            dados = json.load(f)
            for nome, info in dados.items():
                resultado["formularios"][nome] = info.get("formularios", {})
                # resultado["docentes"].append({"nome": nome, "ativo": True})
```

# Disciplinas

```python
disciplinas: list[dict] = json.load(open('turmas.json', 'r',  encoding='utf-8'))
```

```python
for d in disciplinas:
    d['cursos'] = d['curso']
    d['docentes'] = [] # Aproveitando que já está sendo iterado
    del d['curso']
```

## Ajustar as turmas adicionando o campo "docentes"

```python
dictPlinas = {disciplinas[i]['id'] : disciplinas[i] for i in range(len(disciplinas))}
```

```python
for docente, turmasDocente in resultado.get("formularios", {}).items():
    if not turmasDocente:
        continue

    for turma in turmasDocente.keys():
        tDict = dictPlinas.get(turma)
        if tDict is not None:
            tDict.setdefault('docentes', []).append(docente)
```

```python
resultado["disciplinas"] = disciplinas
```

# Docentes e Saldos

```python
import pandas as pd
```

```python
df = pd.read_excel("_ICMC _listagem docentes-somente docentes SME.xls")
```

```python
df.drop(columns=["E-mail", 'Nº USP'])
```

```python
for nome, saldo in zip(df['Nome'], df['saldo']):
    resultado["docentes"].append({'nome': nome, "ativo": True})
    resultado["saldos"].update({nome: saldo})
```

```python
# Se quiser salvar esse resultado em um novo arquivo:
with open('formularios_combinados.json', 'w', encoding='utf-8') as f:
    json.dump(resultado, f, ensure_ascii=False, indent=2)
```
